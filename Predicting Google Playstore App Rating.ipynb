## Handling Missing Values

For dealing with missing values, we're using the following strategy:

   - Random Value Imputation: This method is chosen as it helps to preserve the original distribution of the data.

For the 'Rating' feature, we're using the median value to fill any gaps:

```python
# Function to impute median
def impute_median(series):
    return series.fillna(series.median())

# Applying the imputation function to the 'Rating' column
data['Rating'] = data['Rating'].transform(impute_median)

# Checking the info to ensure no null values exist
print("Dataframe information after median imputation for 'Rating':")
print(data.info())
```

We repeat the process for the 'Size' feature:

```python
# Applying the imputation function to the 'Size' column
data['Size'] = data['Size'].transform(impute_median)

# Checking the info to confirm the imputation
print("Dataframe information after median imputation for 'Size':")
print(data.info())
```

Finally, we address any nulls in the 'Type' feature using the mode:

```python
# Filling missing values in 'Type' with the mode
data['Type'].fillna(str(data['Type'].mode().values[0]), inplace=True)

# Verifying that there are no more missing values
print("Remaining null values after imputations:")
print(data.isnull().sum())
```
